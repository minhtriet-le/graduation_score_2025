# 05 – Build: Chạy toàn bộ pipeline phân tích & xuất chart

Notebook này chạy lần lượt các notebook phân tích bằng `nbconvert`.
Dùng để **tái tạo toàn bộ kết quả** từ đầu (tương đương `make build`).

> Notebook `01_etl` **không** được chạy tự động ở đây để tránh ghi đè Parquet không cần thiết.
> Chạy thủ công `01_etl.ipynb` khi có dữ liệu mới.

```python
import subprocess
import sys
from pathlib import Path

NOTEBOOKS_DIR = Path('.')
NOTEBOOKS = [
    '02_eda.ipynb',
    '03_clustering.ipynb',
    '04_hypothesis.ipynb',
]

for nb in NOTEBOOKS:
    nb_path = NOTEBOOKS_DIR / nb
    print(f'\n{'='*50}')
    print(f'▶ Running: {nb}')
    result = subprocess.run(
        [
            sys.executable, '-m', 'nbconvert',
            '--to', 'notebook',
            '--execute',
            '--inplace',
            '--ExecutePreprocessor.timeout=600',
            str(nb_path),
        ],
        capture_output=True, text=True,
    )
    if result.returncode == 0:
        print(f'✅ Done: {nb}')
    else:
        print(f'❌ Error in {nb}:')
        print(result.stderr[-2000:])
        break

print('\n✅ Build complete! Charts saved to dist/charts/')
```
