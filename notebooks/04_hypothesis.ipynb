# 04 – Kiểm định giả thuyết thống kê

So sánh điểm thi giữa các tỉnh/nhóm bằng:
- **Levene's test** – kiểm tra phương sai đồng nhất
- **Welch's / Student's t-test** – kiểm định trung bình hai mẫu
- **Cohen's d** – effect size
- **Mann-Whitney U** – kiểm định phi tham số

Thêm cặp so sánh mới bất kỳ bằng cách gọi `compare_two_groups()`.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
from src.config import OUTPUT_DIR
from src.plotting import setup_style
from src.stats import compare_two_groups

setup_style()

df = pd.read_parquet('../combined_data.parquet')
print(f'Rows: {len(df):,}')
```
