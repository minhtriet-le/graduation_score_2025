# 03 – Clustering: Phân cụm tỉnh theo điểm thi

Dùng K-Means để phân cụm 63 tỉnh/thành theo điểm thi:
- **Đơn môn:** Toán, Ngữ Văn, Vật lí, Hóa học
- **Đa môn:** Khối Tự Nhiên (Toán + Lý + Hóa), Khối Xã Hội (Văn + Sử + Địa)

Mỗi phân tích gồm: Elbow Method + Silhouette Score để chọn K → Scatter 2D với Convex Hull.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
from src.config import OUTPUT_DIR
from src.plotting import setup_style, kmeans_subject_2d, kmeans_multi_subject_2d

setup_style()

df = pd.read_parquet('../combined_data.parquet')
print(f'Rows: {len(df):,}')
```
